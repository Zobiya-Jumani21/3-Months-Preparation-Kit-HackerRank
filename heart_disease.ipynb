```
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('heart.csv')
df.head()
```

```
df.info()
```

```
df.describe()
#Mean, min, max, std (standard deviation) for each numeric column.
```

```
df.isnull().sum() #for missing values
```

```
df.shape #how many rows and column
```

```
df['target'].value_counts()
```

```
sns.countplot(x='target', data=df)
plt.title('Heart Disease Count (0 = No, 1 = Yes)')
plt.show()
```

```
sns.histplot(data=df, x='age', hue='target', bins=20, kde=True)
plt.title('Age vs Heart Disease')
plt.show()
```

```
sns.countplot(x='sex', hue='target', data=df)
plt.title('Sex vs Heart Disease')
plt.xlabel('Sex (0 = Female, 1 = Male)')
plt.show()
```
